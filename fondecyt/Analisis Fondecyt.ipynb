```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 10),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large',
         'axes.titlepad': 20,
         'axes.titlesize': 24,
         'axes.labelpad': 20,
         'axes.labelsize': 20,
         'lines.linewidth': 3,
         'lines.markersize': 10,
         'xtick.labelsize': 16,
         'ytick.labelsize': 16}
plt.rcParams.update(params)
sns.color_palette("bright")

pd.set_option('display.max_rows', 500)
```

## Fondecyt Postdoctorado

```python
fondecyt = pd.read_csv('data/tabular/fondecyt postdoc 2007-2017.csv')
fondecyt.head()
```

```python
fondecyt=fondecyt.rename(columns = {'año':'Año', 'n_concursados':'Concursados', 'n_aprobados':'Aprobados', 
                                   'monto_solicitado':'Recursos solicitados', 'monto_aprobado':'Recursos aprobados'})
fondecyt['Tasa de aprobación']=np.round(fondecyt['Aprobados']/fondecyt['Concursados']*100,decimals=1)
fondecyt_postdoc=fondecyt.copy()
fondecyt.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax2 = ax.twinx()

fondecyt.plot(kind='line', x='Año', y='Concursados', marker='o', markersize=10, ax=ax, label='Postdoc Concursados')
fondecyt.plot(kind='line', x='Año', y='Aprobados', marker='o', markersize=10, ax=ax, label='Postdoc Aprobados')
ax.set_xlim([2006.5,2017.5])
ax.set_ylim([0,1200])
ax.yaxis.set_ticks(np.arange(0, 1300, 100))
ax.legend(loc=2)

fondecyt.plot(kind='line', x='Año', y='Tasa de aprobación', color='red', grid=True, ax=ax2)
ax2.set_ylim([0,100])
ax2.yaxis.set_ticks(np.arange(0,110,10))
ax2.set_xlim([2006.5,2017.5])
ax2.xaxis.set_ticks(np.arange(2007, 2018, 1))
ax2.grid(linestyle='--')
ax2.legend(loc=1)

ax.set_title('Estadística Proyectos Fondecyt de Postdoctorado')
ax.set_ylabel('Número de proyectos')
ax2.set_ylabel('Tasa (%)')

fig.savefig('figures/estadistica_proyectos_fondecyt_postdoc.pdf')
```

## Fondecyt Postdoctorado y Becas Conicyt de Doctorado

```python
conicyt_doctorado = pd.read_csv('data/tabular/conicyt doctorado 2007-2017.csv')
conicyt_doctorado.head()
```

```python
conicyt_doctorado=conicyt_doctorado.rename(columns = {'año':'Año'})
conicyt_doctorado['Aprobados']=conicyt_doctorado.iloc[:,1:5].sum(axis=1)
conicyt_doctorado.head()
```

```python
mineduc_doctorado=pd.read_csv("data/tabular/mineduc doctorado 2007-2017.csv")
mineduc_doctorado.head()
```

```python
mineduc_doctorado=mineduc_doctorado.rename(columns = {'año':'Año', 'matriculados':'Matriculados'})
mineduc_doctorado.head()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax2 = ax.twinx()
current_palette = sns.color_palette()

fondecyt_postdoc.plot(kind='line', x='Año', y='Concursados', marker='o', markersize=10, ax=ax, label='Postdoc Concursados')
fondecyt_postdoc.plot(kind='line', x='Año', y='Aprobados', marker='o', markersize=10, ax=ax, label='Postdoc Aprobados')
conicyt_doctorado.plot(kind='line', x='Año', y='Aprobados', marker='o', markersize=10, ax=ax, label='Doctorado Aprobados')
ax.set_xlim([2006.5,2017.5])
ax.set_ylim([0,1600])
ax.yaxis.set_ticks(np.arange(0, 1700, 100))
ax.legend(loc=2)

mineduc_doctorado.plot(kind='line', x='Año', y='Matriculados', marker='o', markersize=10, color=current_palette[3], grid=True, ax=ax2, label='Matrícula Doctorado Nacional')
ax2.set_ylim([0,6000])
ax2.set_xlim([2006.5,2017.5])
ax2.xaxis.set_ticks(np.arange(2007, 2018, 1))
ax2.grid(linestyle='--')
ax2.legend(loc=1)

ax.set_title('Estadística Proyectos Fondecyt Postdoc y Becas Doctorado')
ax.set_ylabel('Número de proyectos / becas')
ax2.set_ylabel('Matrícula total Doctorado')

fig.savefig('figures/estadistica_fondecyt_postdoc-conicyt_doctorado.pdf')
fig.savefig('figures/estadistica_fondecyt_postdoc-conicyt_doctorado.png')
```
